```python
def read_csv(path):
    res = {}
    with open(path,"r") as f:
        head = f.readline().strip().split(",")
        for it in f.readlines():
            it = it.strip().split(",")
            if len(head) == len(it):
                for _h, _i in zip(head, it):
                    res[_h] = res.get(_h, [])
                    res[_h].append(_i)
            else:
                print("error")
    return res
```
